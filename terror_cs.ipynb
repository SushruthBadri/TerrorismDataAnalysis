```python
#write a function to label an incident that was both successful and suicidal
def labelincident(tdf,eventid):
    dict1 =tdf[tdf['eventid']==eventid]
    print(dict1['success'],':',dict1['suicide'])
    if (int(dict1['success'])==1) & (int(dict1['suicide'])==1):
        return True
    else:
        return False
```

```python
eid = 200502110001
labelincident(tdf,eid)
```

```python
#Create a new category representing if the incident occured in Afghanistan, Pakistan or India as one level of the category and all the other countries as another level
countries = list(tdf['country_txt'])
category = []
countrytrue = ['India','Pakistan','Afghanistan']
for country in countries:
    if country in countrytrue:
        category.append('1')
    else:
        category.append('0')
tdf['category'] = category
tdf.head()
```

```python
#How many incidents happened in Af-Pak-India vs ROW?
print('ROW incidents:', tdf[tdf['category']=='0']['category'].size,'afg-pak-ind incidents:', tdf[tdf['category']=='1']['category'].size)
```

```python
#List the number of suicides attacks and average kills by Af-Pak-India vs ROW. Rename columns in the output as Average_Kills and Number_Incidents.
tdf.groupby('category').agg({'suicide':'count','nkill':'mean'}).rename(columns={'suicide':'Number_Incident','nkill':'Average_Kills'})
```
